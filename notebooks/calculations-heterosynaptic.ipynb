Here we use symbolic methods to check the eigenvalue calculation in SI Appendix, S3

```python
from sympy import *
init_printing()
```

```python
lmb = symbols('lambda')
k_ltd, k_pf, k_mf, k_cf, beta_l, w_pc, k_e, g_t, w_h, v, k_v, sigma_h2, tau_w = symbols('k_ltd, k_pf, k_mf, k_cf, beta_l, w_pc, k_e, g_t, w_h, v, k_v, sigma_h2, tau_w')
```

```python
mf_pf = symbols('mf_pf')
```

```python
g = k_e*(k_mf*v - w_pc*k_pf*w_h)
```

```python
dxdt = Matrix([1/tau_w*(-w_h - k_ltd*k_pf*k_cf*beta_l*(g_t - g)*sigma_h2), 
               -k_v * mf_pf * w_h])
```

```python
J = dxdt.jacobian(Matrix([w_h, v]))
```
